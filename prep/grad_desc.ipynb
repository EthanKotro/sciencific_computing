```python
import numpy as np
import pandas as pd

def gradient_descent( x0, alpha, n):
    def f(x):
        return x**2 - 6*x + 9

    def df(x):
        return 2*x - 6
    """
    Performs gradient descent to minimize a function.

    Args:
        f (callable): The function to minimize.
        df (callable): The derivative of the function.
        x0 (float): The initial guess for x.
        alpha (float): The learning rate.
        n (int): The number of iterations.

    Returns:
        tuple: A tuple containing two lists:
               - x_values: A list of x-values at each iteration.
               - f_values: A list of f(x)-values at each iteration.
    """
    x = x0
    x_values = [x]
    f_values = [f(x)]
    for _ in range(n):
        gradient = df(x)
        x = x - alpha * gradient
        x_values.append(x)
        f_values.append(f(x))
    return x_values, f_values

# Define the function and its derivative


```

```python
x0 = 2
alpha = 1.0
n = 10

x_vals, f_vals = gradient_descent( x0, alpha, n)

# Create a table using pandas
data = {'iteration': range(n + 1), 'x-value': x_vals, 'f(x)-value': f_vals}
df_output = pd.DataFrame(data)
print(df_output)
```
